````python
df_salary = pd.read_csv('salarios.csv')
df_salary = df_salary.loc[:,['Player', 'Weekly Salary']]
df_wage = df_players.merge(df_salary, right_on='Player', left_on='Name', how='inner')
df_wage['Weekly Salary'] = df_wage['Weekly Salary'].astype(int)
````

````python
wageCT = ColumnTransformer(
    transformers=[
        (
            "OneHotEncoder",
            OneHotEncoder(drop='first', handle_unknown='ignore'),
            [
                'Nationality',
                "Preffered_Foot",
                "Skill_Moves",
                "Work_Rate",
                "Weak_foot",
                'Club_Position'
            ],
        ),
        (
            "StandardScaler",
            StandardScaler(),
            [
                "Weight",
                "Height",
                'Age',
                "Aggression",
                "Reactions",
                "Vision",
                "Composure",
                "Acceleration",
                "Speed",
                "Balance",
                "Strength",
                "Agility",
                "Jumping",
                "Finishing",
                "Volleys",
                "Penalties"
            ],
        ),
        (
            "MinMaxScaler",
            MinMaxScaler(),
            [
                "Ball_Control",
                "Dribbling",
                "Marking",
                "Sliding_Tackle",
                "Standing_Tackle",
                "Interceptions",
                "Crossing",
                "Short_Pass",
                "Long_Pass",
                "Stamina",
                "Heading",
                "Shot_Power",
                "Curve",
                "Freekick_Accuracy",
                "Long_Shots"
            ],
        ),
    ]
)


wage_pipe = Pipeline(
    [
        ("preprocessing", wageCT),
        ("classification", RandomForestRegressor()),
    ]
)
````

````python
y = df_wage.loc[:, "Weekly Salary"]
X = df_wage.drop(['Player', 'Weekly Salary', 'Name', 'Seleccionado'], axis=1)

display(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)

pipe = wage_pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

print(f"R^2: {round(r2_score(y_test, y_pred), 3)}")
````
