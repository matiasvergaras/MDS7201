### 1.3.4 Tabla de Resumen [1 punto]

Para generar la tabla de resumen:

- [ ] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [ ] Hacer una simplificación a través de un mapeo (investigar el método `map()`) de la variable `parental level of education`  según la siguiente conversión: 

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas. 
- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [ ] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'. 


Utilizar la tabla de resultados esperados como guía para desarrollar este punto.

````python
def alumnos_promedio(dataframe_in):
    dataframe_in['GPA'] = (dataframe_in['math score'] + dataframe_in['writing score'] + dataframe_in['reading score'])/3
    return dataframe_in


def alumnos_educacion_simple(dataframe_in):
    map_dict = {'some high school': 'school',
                'some college': 'school',
                'high school': 'school',
                "bachelor's degree": 'college',
                "associate's degree": 'college',
                "master's degree": 'postgraduate'}
    dataframe_in["simple parental level of education"] = dataframe_in["parental level of education"].map(map_dict)
    return dataframe_in


def alumnos_porcentaje(dataframe_in):
    total = dataframe_in["count"].sum()
    dataframe_in["porcentaje"] = (100*dataframe_in["count"]/total).round(2)
    dataframe_in["porcentaje"] = dataframe_in["porcentaje"].astype(str) + " %"
    return dataframe_in

df_grades = alumnos_promedio(df_grades)
df_grades = alumnos_educacion_simple(df_grades)
df_grades = df_grades.set_index(['race/ethnicity', 'simple parental level of education'])

df_promedio = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).mean().round(2)
df_promedio["count"] = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).size()
df_porcentaje = alumnos_porcentaje(df_promedio)
display(df_porcentaje)
````

````python
other_grades = pd.read_csv("other_grades.csv", index_col="names")
display(other_grades)

df_grades_outer = df_grades.set_index('names').join(other_grades, how='outer')
display(df_grades_outer)

df_grades_left = df_grades.set_index('names').join(other_grades, how='left')
display(df_grades_left)

df_grades_right = df_grades.set_index('names').join(other_grades, how='right')
display(df_grades_right)

df_grades_inner = df_grades.set_index('names').join(other_grades, how='inner')
display(df_grades_inner)
````

```
Recordando que la cantidad correcta de entradas (la cantidad de alumnos) es 875, podemos notar de inmediato que las únicas operaciones que respetan esta condición al utilizar los nombres de los estudiantes para la intersección son inner join y left join, por lo cual son las únicas aptas para generar el nuevo DataFrame. Sin embargo, tratemos de entender el resultado de cada operación:

Inner join: corresponde a una intersección de los dataframes, donde sobreviven solo los elementos que están en ambos. Dado que df_grades contiene 875 entradas y que other_grades incluye estas más algunos duplicados, la intersección resulta en solamente las 875 filas correspondientes (pero ahora incorporando las columnas nuevas). 

Left join: corresponde a la unión de la intersección entre ambos dataframes con el dataframe de la izquierda, completando con NaN aquellas columnas que provengan del dataframe de la derecha pero que no estén en la intersección con el de la izquierda. Dado que en este caso el dataframe de la izquierda es df_grades y el de la derecha other_grades, no caemos en este caso y por ende solo se agregan las nuevas columnas sin generar valores vacíos.

Right join: lo mismo que Left join, pero al revés. Manda el dataframe de la derecha, y se completan con valores NaN aquellas columnas provenientes del dataframe de la izquierda pero para las cuales el dataframe de la derecha no tiene información. En este caso no sirve, pues el dataframe de la derecha - other_grades - tiene alumnos duplicados, que se mantienen al hacer el join y se completan además con valores NaN.

Outer join: Corresponde a la unión menos la intersección de los dataframes. En este caso, como el dataset de la derecha incluye las filas duplicadas, la unión también lo hace y la intersección no lo corrige. Como resultado, se mantienen las 1000 observaciones y con muchos valores vacios en la tabla final.
```
