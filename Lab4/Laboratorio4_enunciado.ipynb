### 1.3.4 Tabla de Resumen [1 punto]

Para generar la tabla de resumen:

- [ ] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [ ] Hacer una simplificaci칩n a trav칠s de un mapeo (investigar el m칠todo `map()`) de la variable `parental level of education`  seg칰n la siguiente conversi칩n: 

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [ ] Agregar seg칰n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas. 
- [ ] Agregar seg칰n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [ ] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'. 


Utilizar la tabla de resultados esperados como gu칤a para desarrollar este punto.

````python
def alumnos_promedio(dataframe_in):
    dataframe_in['GPA'] = (dataframe_in['math score'] + dataframe_in['writing score'] + dataframe_in['reading score'])/3
    return dataframe_in


def alumnos_educacion_simple(dataframe_in):
    map_dict = {'some high school': 'school',
                'some college': 'school',
                'high school': 'school',
                "bachelor's degree": 'college',
                "associate's degree": 'college',
                "master's degree": 'postgraduate'}
    dataframe_in["simple parental level of education"] = dataframe_in["parental level of education"].map(map_dict)
    return dataframe_in


def alumnos_porcentaje(dataframe_in):
    total = dataframe_in["count"].sum()
    dataframe_in["porcentaje"] = (100*dataframe_in["count"]/total).round(2)
    dataframe_in["porcentaje"] = dataframe_in["porcentaje"].astype(str) + " %"
    return dataframe_in

df_grades = alumnos_promedio(df_grades)
df_grades = alumnos_educacion_simple(df_grades)
df_grades = df_grades.set_index(['race/ethnicity', 'simple parental level of education'])

df_promedio = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).mean().round(2)
df_promedio["count"] = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).size()
df_porcentaje = alumnos_porcentaje(df_promedio)
display(df_porcentaje)
````

Outer join: Corresponde a la uni칩n menos la intersecci칩n de los dataframes. En este caso, como el dataset de la derecha incluye las filas duplicadas, la uni칩n tambi칠n lo hace y la intersecci칩n no lo corrige.
